```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("newbig.csv")
```

```python
df = df.transpose()
df = df.drop("Unnamed: 0")
display(df)
```

```python
df["Win Percentage"] = np.diag(df)
display(df)
```

```python
sort_df = df[["Win Percentage"]].sort_values("Win Percentage", ascending=False)
pd.set_option('display.max_rows', 200)
display(sort_df.head(200))
```
