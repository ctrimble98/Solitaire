```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../output/heuristic-test-results.csv", skiprows=1)
df = df.transpose()
df = df.drop("Unnamed: 0")
df.index = df.index.str[1:]
df["Win Percentage"] = np.diag(df)
display(df.head())
```

```python
sort_df = df[["Win Percentage"]].sort_values("Win Percentage", ascending=False)
pd.set_option('display.max_rows', 100)
display(sort_df.head(100))
```
